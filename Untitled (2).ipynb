```python
import xarray as xr
import xesmf as xe
```

```python
# Open dataset 
dr = '/home/jovyan/SOARS_2021/Data/CESM2/'
fn = 'b.e11.B20TRC5CNBDRD_no_pinatubo.f09_g16.011.pop.h.photoC_diaz.199001-200512.nc?dl=0'

data = xr.open_dataset(dr+fn)
data
```

```python
TAREA = data.TAREA
TAREA
```

```python
TAREA_m2 = TAREA / (100**2)
TAREA_m2
```

```python
TAREA = TAREA_m2.rename({'nlat':'lat',
                         'nlon':'lon'})
print(TAREA)
```

```python
# to re-grid
ds_out = xe.util.grid_global(1,1)
regridder = xe.Regridder(TAREA, ds_out, 'bilinear')
dr_out = regridder(TAREA)
```

```python
dr_out.to_netcdf('TAREA_180x360_06/22/21.nc')
```
